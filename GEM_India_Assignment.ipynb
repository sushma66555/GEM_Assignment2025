```python
import json

# List to store JSON objects
json_list = []

for i, row in df.iterrows():
    json_obj = {
        "country_name": row['country_name'],
        "Year": int(row['Year']),
        "TEA": int(row['Total early-stage Entrepreneurial Activity']),
        "Fear_of_failure": float(row['Fear of failure']),
        "Derived_Opportunity_Perception": float(row['Derived_Opportunity_Perception'])
    }
    json_list.append(json_obj)

# Check first 5 JSON objects
print(json.dumps(json_list[:5], indent=2))
```

```python
with open("GEM_India_2021.json", "w") as f:
    json.dump(json_list, f, indent=2)
```

```python
# Required keys for validation
required_keys = ["country_name", "Year", "TEA", "Fear_of_failure", "Derived_Opportunity_Perception"]

# Validate each JSON object
for idx, item in enumerate(json_list):
    missing_keys = [key for key in required_keys if key not in item]
    if missing_keys:
        print(f"Row {idx} missing keys: {missing_keys}")
```
